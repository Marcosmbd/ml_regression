Importando módulos

```python
from pycaret.datasets import get_data
from pycaret import regression
```

Definindo o dataset para predição e o Target, que no caso será o Preço de venda. <br>
Com poucas linhas, o Pycaret já defina a base de treino e teste, além de nos dar uma descrição bem completa dos dados, <br>
além de já tratar as colunas categórias, transformando em numéricas.

```python
df = get_data('house')
regression.setup(df, target=('SalePrice'))
```

Agora vamos criar nossa máquina preditora mas inicialmente, vamos ver a que mais se sai melhor.

```python
regression.compare_models()
```

Vemos que o algoritmo que mais se saiu melhor foi o "Light Gradient Boosting Machine" então vamos selecioná-lo <br> 
e criar nossa máquina Preditora. A função tune_model ajusta automaticamente os hiperparâmetros do algoritmo. <br> 
Ao criar o pycaret já faz o fit com a base de treino, exibindo os resultados.

```python
LGBM = regression.create_model('lightgbm')
LGBM = regression.tune_model(LGBM, choose_better=True)
```

Vamos agora ver os resultados na base de teste.<br>
A coluna "prediction_label", é o resultado da nossa predição

```python
regression.predict_model(LGBM)
```

Para finalizar, vamos ver o resultado com a base completa, e a assertividade de forma mais visual, <br>
usando a função evaluate_model e finalizar a ML.

```python
regression.predict_model(LGBM, data=df)
```

```python
regression.evaluate_model(LGBM)
```

Para salvar sua ML, use a função regression.save_model() e para carregar use regression.load_model()

```python
regression.finalize_model(LGBM)
```
